```python
def retrieve(query, k=5):
    query_vec = embed_model.encode([query])
    distances, indices = index.search(query_vec, k)
    results = df.iloc[indices[0]].copy()
    results["distance"] = distances[0]
    return results
```

```python
def build_prompt(query, k=5):
    results = retrieve(query, k)
    context = ""
    for _, row in results.iterrows():
        context += f"[{row['id']}] ({row['source']}): {row['translation']}\n"
    
    prompt = f"""You are a scripture citation assistant. Answer using ONLY the verses below.

STRICT FORMAT: Write your answer as a bulleted list. Every single bullet point MUST end with a verse ID in brackets. Do not write any sentence without a citation.

Example format:
- One should perform duty without attachment to outcomes [GITA_2_47]
- Attachment to results causes suffering [GITA_3_19]

If the verses do not contain enough information, respond with exactly: "The provided verses do not directly answer this question."

Verses:
{context}
Question: {query}
Answer:"""
    return prompt
```

```python
import torch
from transformers import AutoTokenizer, AutoModelForCausalLM

GEMMA_PATH = "/kaggle/input/models/google/gemma/transformers/2b-it/3"

tokenizer = AutoTokenizer.from_pretrained(GEMMA_PATH)

gemma_model = AutoModelForCausalLM.from_pretrained(
    GEMMA_PATH,
    torch_dtype=torch.bfloat16,
    device_map="cuda"
)

print("Gemma loaded successfully")
print("Device:", gemma_model.device)
```

```python
def generate_answer(query, k=5, max_new_tokens=300):
    prompt = build_prompt(query, k=k)

    chat = [{"role": "user", "content": prompt}]
    inputs = tokenizer.apply_chat_template(
        chat,
        add_generation_prompt=True,
        return_tensors="pt",
        return_dict=True
    ).to(gemma_model.device)

    outputs = gemma_model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=False
    )

    response = tokenizer.decode(
        outputs[0][inputs["input_ids"].shape[-1]:],
        skip_special_tokens=True
    )
    return response

print("generate_answer() defined")
```

```python
import re

def check_hallucination(query, k=5):
    results = retrieve(query, k)
    answer = generate_answer(query, k=k)
    
    # get retrieved IDs directly from the dataframe (ground truth), not by scanning the prompt text
    retrieved_ids = set(results['id'])
    
    # extract verse IDs Gemma cited in its answer
    cited_ids = set(re.findall(r'[\[\(]([A-Z]+_[A-Za-z0-9_]+)[\]\)]', answer))
    
    fake_ids = cited_ids - retrieved_ids
    
    print("Retrieved IDs:", retrieved_ids)
    print("Cited IDs:    ", cited_ids)
    print("HALLUCINATED IDs (cited but not retrieved):", fake_ids if fake_ids else "None ✅")
    print()
    return fake_ids

for source, q in test_queries.items():
    print(f"===== {source} =====")
    check_hallucination(q, k=5)
```

```python
def retrieve_per_source(query, k_per_source=2):
    query_vec = embed_model.encode([query])
    
    sources = ['Gita', 'Quran', 'Guru Granth Sahib', 'Bible']
    all_results = []
    
    for src in sources:
        # get only rows for this source
        subset = df[df["source"] == src].reset_index(drop=True)
        subset_texts = subset["translation"].tolist()
        
        # embed just this subset (simple, no separate FAISS index needed yet)
        subset_embeddings = embed_model.encode(subset_texts, show_progress_bar=False)
        
        # brute-force distance search using numpy (fine for small subsets)
        import numpy as np
        distances = np.linalg.norm(subset_embeddings - query_vec, axis=1)
        top_k_idx = np.argsort(distances)[:k_per_source]
        
        top_rows = subset.iloc[top_k_idx].copy()
        top_rows["distance"] = distances[top_k_idx]
        all_results.append(top_rows)
    
    return pd.concat(all_results, ignore_index=True)
```

```python
def build_comparison_prompt(query, k_per_source=2):
    results = retrieve_per_source(query, k_per_source=k_per_source)
    
    context = ""
    for src in ['Gita', 'Quran', 'Guru Granth Sahib', 'Bible']:
        src_rows = results[results["source"] == src]
        context += f"\n{src} verses:\n"
        for _, row in src_rows.iterrows():
            context += f"[{row['id']}] {row['translation']}\n"
    
    prompt = f"""You are a comparative scripture assistant. Compare how different traditions address the topic below, using ONLY the verses listed under "Verses:" below.

STRICT FORMAT: For EACH tradition (Gita, Quran, Guru Granth Sahib, Bible), write exactly one bullet point summarizing its view, ending with the real verse ID(s) from the Verses section, in brackets.

Example format (structure only — do NOT reuse these IDs, they are not real):
- [Tradition]: [summary of that tradition's view] [PLACEHOLDER_ID]

IMPORTANT RULES:
- Only cite verse IDs that literally appear in the "Verses:" section below.
- Never reuse GITA_2_47, QURAN_1_3, GGS_914_15, or BIBLE_Psalms_82_3 unless they actually appear in the Verses section below for this specific query.
- If a tradition's verses do not relate to the topic, write: "- [Tradition name]: The provided verses do not directly address this topic."

Verses:
{context}
Topic: {query}
Answer:"""
    return prompt
```

```python
print(build_comparison_prompt("duty and compassion", k_per_source=2))
```

```python
def generate_comparison(query, k_per_source=2, max_new_tokens=400):
    prompt = build_comparison_prompt(query, k_per_source=k_per_source)

    chat = [{"role": "user", "content": prompt}]
    inputs = tokenizer.apply_chat_template(
        chat,
        add_generation_prompt=True,
        return_tensors="pt",
        return_dict=True
    ).to(gemma_model.device)

    outputs = gemma_model.generate(
        **inputs,
        max_new_tokens=max_new_tokens,
        do_sample=False
    )

    response = tokenizer.decode(
        outputs[0][inputs["input_ids"].shape[-1]:],
        skip_special_tokens=True
    )
    return response

print("generate_comparison() defined")
```

```python
import re

def check_hallucination_comparison(query, k_per_source=2):
    results = retrieve_per_source(query, k_per_source=k_per_source)
    answer = generate_comparison(query, k_per_source=k_per_source)
    
    retrieved_ids = set(results['id'])
    
    # find all bracket/paren groups, then split multi-ID groups by comma
    raw_groups = re.findall(r'[\[\(]([^\[\]\(\)]+)[\]\)]', answer)
    cited_ids = set()
    for group in raw_groups:
        for piece in group.split(","):
            piece = piece.strip()
            if re.match(r'^[A-Z]+_[A-Za-z0-9_]+$', piece):
                cited_ids.add(piece)
    
    fake_ids = cited_ids - retrieved_ids
    
    print("Retrieved IDs:", retrieved_ids)
    print("Cited IDs:    ", cited_ids)
    print("HALLUCINATED IDs (cited but not retrieved):", fake_ids if fake_ids else "None ✅")
    print()
    return fake_ids
```

```python
comparison_test_topics = [
    "duty and compassion",
    "forgiveness",
    "the nature of the soul",
    "charity"
]

for topic in comparison_test_topics:
    print(f"===== {topic} =====")
    print(generate_comparison(topic, k_per_source=2))
    check_hallucination_comparison(topic, k_per_source=2)
```

```python
def ask_scripture_engine(query, mode):
    if mode == "Single Scripture Answer":
        return generate_answer(query, k=5)
    else:
        return generate_comparison(query, k_per_source=2)

print("ask_scripture_engine() defined")
```

```python
import gradio as gr

demo = gr.Interface(
    theme=gr.themes.Glass(primary_hue="rose"),
    fn=ask_scripture_engine,
    inputs=[
        gr.Textbox(label="Enter your question or topic", placeholder="e.g. What is Dharma?"),
        gr.Radio(["Single Scripture Answer", "Cross-Tradition Comparison"], 
                  label="Mode", value="Single Scripture Answer")
    ],
    outputs=gr.Textbox(label="Answer with Citations", lines=12),
    title="Gemma Scripture Engine",
    description="Ask a spiritual/philosophical question and get grounded answers with verse citations from the Gita, Quran, Guru Granth Sahib, and Bible."
)

print("Gradio interface built")
```
